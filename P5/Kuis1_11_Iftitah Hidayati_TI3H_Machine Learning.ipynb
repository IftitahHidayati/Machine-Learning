```python
# =================================================
# 2. Distribusi peluang tagihan di tiap-tiap region
# =================================================
sorting_by_region = dataset.groupby(['region'], as_index=False).count()
print('=================================================')
print('Berikut nama-nama region yang ada pada dataset')
print('=================================================')
print(sorting_by_region['region'])

total_charges = 0
northeast_charges = 0
northwest_charges = 0
southeast_charges = 0
southwest_charges = 0
x = len(dataset)
# Mengecek pembagian region
for i in range(0, x):
  total_charges += dataset.iloc[i][6]
  if (dataset.iloc[i][5] == 'northeast'):
    northeast_charges += dataset.iloc[i][6]
  if (dataset.iloc[i][5] == 'northwest'):
    northwest_charges += dataset.iloc[i][6]
  elif (dataset.iloc[i][5] == 'southeast'):
    southeast_charges += dataset.iloc[i][6]
  elif (dataset.iloc[i][5] == 'southwest'):
    southwest_charges += dataset.iloc[i][6]
print('=================================================')
a = southeast_charges/total_charges
b = southwest_charges/total_charges
c = northeast_charges/total_charges
d = northwest_charges/total_charges
print('Peluang region southeast adalah ',a)
print('Peluang region southwest adalah ',b)
print('Peluang region northeast adalah ',c)
print('Peluang region northwest adalah ', d)
print('=================================================')
print('Total peluang (test) = ',a+b+c+d)
```

```python
# ===============================================================================================
# 5. Mana yang lebih mungkin terjadi
# a. Seseorang perokok dengan BMI diatas 25 mendapatkan tagihan kesehatan diatas 16.7k, atau

# b. Seseorang non perokok dengan BMI diatas 25 mendapatkan tagihan kesehatan diatas 16.7k
# ===============================================================================================
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()

dataset['smoker'] = encode.fit_transform(dataset['smoker'])

dataset.head()
```

```python
total = 0
x = len(dataset)
for i in range(0, x):
    if(dataset.iloc[i][4] == 1):
        if (dataset.iloc[i][2] >= 25):
              total += 1
              
total1 = 0
x = len(dataset)
for i in range(0, x):
    if(dataset.iloc[i][4] == 0):
        if (dataset.iloc[i][2] >= 25):
              total1 += 1

print('Jadi terdapat ',total,' yang merupakan perokok dan mendapat tagihan 16,7k diatas BMI 25')
print('Jadi terdapat ',total1,' yang merupakan non perokok mendapat tagihan 16,7k diatas BMI 25')

keseluruhan = total1 + total
probabilitas_perokok = total / keseluruhan
probabilitas_nonperokok = total1 / keseluruhan

print('-----------------------------------------------------------------------------------')

print('jadi probabilitas untuk perokok dengan tagihan 16,7k diatas BMI 25 adalah: ', probabilitas_perokok)
print('jadi probabilitas untuk non perokok dengan tagihan 16,7 k diatas BMI 25 adalah: ', probabilitas_nonperokok)

print('-----------------------------------------------------------------------------------')

print('jadi yang paling memungkinkan adalah Seseorang yang non perokok dengan tagihan 16,7 k diatas BMI 25' )
```
